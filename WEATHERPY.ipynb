```python
import openweathermapy.core as owm


from config import api_key

url = "http://api.openweathermap.org/data/2.5/forecast" \
        "?APPID=" + api_key
    
units = "Imperial"
settings = {"units": units, "appid" :api_key}
```

```python
cities = []
while (len(cities) < 700):
    lat, lon = rnd.uniform(-180,180), rnd.uniform(-90, 90)
    city = citipy.nearest_city(lat, lon)
    if (city not in cities):
        cities.append(city)
for i in (range(700)):
    print(cities[i].city_name)
    
```

```python
query_url = f"{url}&units={units}&q="


city_weather = []

i = 0

for city in cities:
    i = i + 1
    weather_dict = {}
    print ("Processing Record", i, "of Set 1 | " + city.city_name)
    final_url = query_url + city.city_name
    print (final_url)
   
    response = requests.get(final_url).json()
    if response['cod'] == "200":
        weather_dict["city"] = response['city']['name']
        weather_dict["cloudiness"] = response['list'][0]['clouds']['all']
        weather_dict["country"] = response['city']['country']
        weather_dict["date"] = response['list'][0]['dt']
        weather_dict["humidity"] = response['list'][0]['main']['humidity']
        weather_dict["lat"] = response['city']['coord']['lat']
        weather_dict["lng"] = response['city']['coord']['lon']
        weather_dict["max temp"] = response['list'][0]['main']['temp_max']
        weather_dict["wind speed"] = response['list'][0]['wind']['speed']
        city_weather.append(weather_dict)
```

```python
city_weather_df = pd.DataFrame(city_weather).head(500)

city_weather_df
```

```python
# Your objective is to build a series of scatter plots to showcase the following relationships:

# Temperature (F) vs. Latitude
# Humidity (%) vs. Latitude
# Cloudiness (%) vs. Latitude
# Wind Speed (mph) vs. Latitude

# Build a scatter plot for each data type
plt.scatter(city_weather_df["max temp"], 
            city_weather_df["lat"],
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Zip Codes")

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temperature (01/05/17)")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)


# Show plot
plt.show()
```
